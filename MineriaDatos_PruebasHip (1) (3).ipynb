# II. Pruebas de hipótesis

Una __hipótesis estadística__ es una conjetura acerca del parámetro de una población o de su distribución. Si la hipótesis estadística especifica completamente a la distribución se le conoce como _hipótesis simple_; en caso contrario se le llama _hipótesis compuesta_. Cuando se plantea una prueba de hipótesis, las dos hipótesis complementarias se llaman __hipótesis nula ($H_0$)__ e __hipótesis alternativa ($H_1$)__.
El objetivo de un problema de pruebas de hipótesis es indicar, a partir de una muestra de la población, cual de las dos hipótesis es verdadera. 

1. Hipótesis simple
$$ X \sim Poi(15) $$
$$ Y \sim Ber(0.4)$$

2. Hipótesis compuesta
$$ X \sim N(0, \sigma^2), \sigma^2 > 0 $$
$$ T \sim Exp(\lambda), \lambda > 0 $$

Una __prueba estadística $\gamma$__ es una regla que especifica:

i. Para qué valores de la muestra la decisión es aceptar $H_0$ como verdadera.

ii. Para qué valores de la muestra la decisión es rechazar $H_0$ como verdadera. 

## Error tipo I y Error tipo II

Al subconjunto del espacio muestral $\mathbf{X}$ denotado como $C_\gamma$ para el cual se rechaza $H_0$ se le conoce como __región crítica__ o __región de rechazo__.

En un problema de contraste de hipótesis $H_0: \theta \in \Theta_0$ vs $H_1: \theta \in \Theta_1$ pueden existir dos tipos de error: 
1. Error tipo I. Rechazar $H_0$ dado que es verdadera.
2. Error tipo II. Aceptar $H_0$ dado que es falsa.

Y podemos definir el tamaño de cada uno de los errores de la siguiente manera:
1. Tamaño del error tipo I ($\alpha$) $= \mathbb{P}[\text{Rechazar } H_0 | H_0 \text{ es verdadera}]$
2. Tamaño del error tipo II ($\beta$) $ = \mathbb{P}[\text{Aceptar} H_0 | H_0 \text{ es falsa}]$

## Pruebas paramétricas

En este tipo de pruebas nos interesa probar que un parámetro de interés es igual, mayor, menor o distinto a cierto valor, suponiendo una distribución conocida de la población. 
$$ H_0: \theta = \theta_0 \text{ vs } H_1: \theta \not = \theta_0 $$

Para obtener la región crítica y la regla de decisión en estas pruebas podemos utilizar el __Lema de Neyman-Pearson__ o el __Cociente de verosimilitudes generalizado__. 

### Lema de Neyman - Pearson

Sea $X_1,...X_n$ muestra aleatoria de una población con fdp $f(\underline{X}; \theta)$ con $\theta$ que puede tomar valores $\theta_0$ o $\theta_1$. Sea $\alpha$ fijo, $0 < \alpha < 1$. Entonces la prueba estadística más potente de tamaño $\alpha$ para contrastar la hipótesis $H_0: \theta = \theta_0 \text{ vs } H_1: \theta = \theta_1 $ cumple que:
1. $\mathbb{P}[x \in C | \theta_0] = \alpha$
2. $\lambda(x) \leq k$, si $x \in C$. $$\lambda(x) = \frac{L(\theta_0; x)}{L(\theta_1;x)} \leq k $$
Es decir, rechazamos $H_0$ si $\lambda(x) \leq k$.

En otras palabras, dado el valor de $\alpha$ fijo y el contraste de hipótesis simple _vs_ simple, el Lema de Neyman-Pearson nos permite encontrar la mejor región crítica y la regla de rechazo para la hipótesis nula. 

_Nota_: El Lema de NP puede utilizarse en un contraste de simple _vs_ compuesta, donde el valor de $\theta_1$ podemos suponerlo menor o mayor al valor de $\theta_0$.

## P-value

Dado un contraste de hipótesis, la prueba estadística $\gamma$ y la región de rechazo $C_\gamma$, el _p-value_ (también llamado valor-p o p-valor) se define como la probabilidad de obtener un valor al menos tan extremo como el estadístico observado. Es decir, el _p-value_ es la probabilidad que se obtiene de suponer la hipótesis nula como verdadera, obtener un estadístico con respecto a los datos y posteriormente calcular la probabilidad de que el estadístico observado sea al menos tan extremo como el estadístico obtenido. 

__Regla de decisión:__ Rechazar $H_0$ al nivel de significancia $\alpha$ si $p-value < \alpha$. 

__Ejemplo 1:__ Un estudio de médico reveló que la probabilidad de padecer hipertensión en México es de un cuarto. Se tomaron 120 personas al azar para preguntarles si tienen o no hipertensión, de las cuales 35 personas respondieron que sí. Plantee el contraste de hipótesis que permita concluir si el estudio cuenta con pruebas estadísticas que avalen dicha probabilidad planteada. Considera $\alpha = 0.05$.

__Solución__

Ya que las personas pueden tener o no hipertensión, podemos pensar esto como una variable aleatoria con distribución Bernoulli, donde el contraste de hipótesis planteado es
$$H_0: \theta = \frac{1}{4} \text{ vs } H_1: \theta = \theta_1 \text{ , } \theta_1 > \frac{1}{4}$$

Utilicemos el Lema de Neyman-Pearson:

Sabemos que
$$L(\theta;x) = f(\underline{X}; \theta) = \prod_{i=1}^n \theta^{x_i} (1-\theta)^{1-x_i} = \theta^{\sum x_i} (1 - \theta)^{n - \sum x_i} = (1 - \theta)^n \left( \frac{\theta}{1-\theta} \right)^{\sum x_i}$$

Entonces,
$$\lambda(x) = \frac{L(\theta_0; x)}{L(\theta_1;x)} = \frac{(1 - \theta_0)^n \left( \frac{\theta_0}{1-\theta_0} \right)^{\sum x_i}}{(1 - \theta_1)^n \left( \frac{\theta_1}{1-\theta_1} \right)^{\sum x_i}} = \left(\frac{1-\theta_0}{1-\theta_1} \right)^n \left(\frac{\theta_0 (1-\theta_1)}{\theta_1 (1-\theta_0)} \right)^{\sum x_i} \leq k$$

$$ \left(\frac{\theta_0 (1-\theta_1)}{\theta_1 (1-\theta_0)} \right)^{\sum x_i} \leq k \left(\frac{1-\theta_0}{1-\theta_1} \right)^{-n} $$

$$\sum x_i \ln\left(\frac{\theta_0 (1-\theta_1)}{\theta_1 (1-\theta_0)} \right) \leq k_1$$

Como $\theta_0 = \frac{1}{4} < \theta_1$, 
$$\sum x_i \geq k_1 \ln\left(\frac{\theta_0 (1-\theta_1)}{\theta_1 (1-\theta_0)} \right)^{-1} = k^* $$

Dado lo anterior, rechazamos $H_0$ si $\sum x_i  \geq k_2$. 

No obstante, sabemos que $X \sim Ber(\theta)$, y bajo la hipótesis nula, $X \sim Ber(\frac{1}{4})$. Luego, $\sum x_i \sim Bin(120, \frac{1}{4})$. Entonces, 
$$\mathbb{P}\left[\sum x_i \geq k^* \Big | \sum x_i \sim Bin(120, \frac{1}{4}) \right] = 0.05$$
donde $k^*$ es el cuantil $0.95$ de una $Bin(120, \frac{1}{4})$.

Por lo tanto, $\gamma$: Rechazar $H_0$ si $\sum x_i \geq q_{0.95} = 38$. 

Sabemos que $\sum x_i = 35 < 38$. Por lo tanto no existen pruebas estadísticas suficientes para rechazar $H_0$. Concluimos que la probabilidad de sufrir hipertensión es de 0.25.

Hagámoslo por $p-value$. Sabemos que $T = \sum x_i \sim Bin(120, \frac{1}{4})$, y el valor observado es $\sum x_i = 35$. Entonces $$p-value = \mathbb{P}\left[ T \geq 35 \Big | T = Bin(120, \frac{1}{4}) \right] = 0.17086 $$
Como $p-value = 0.17086 > 0.05$ no existen pruebas estadísticas suficientes para rechazar $H_0$.

```r
qbinom(0.95,120,0.25) ## Cuantil 1-a de una bin(120,0.25)

1 - pbinom(34, 120, 0.25) ##Proba de T >= 35
```

## Prueba de Kolmogorov-Smirnov

La prueba de KS es una prueba de __bondad de ajuste__ que permite probar si una muestra aleatoria sigue una distribución hipotética. $$H_0: F_X(x) = F^*(x) \text{ vs } H_1: F_X(x) \not = F^*(x)$$
donde $F_X(x)$ es la distribución de la muestra (es desconocida) y $F^*(x)$ es una distribución hipotética. 

Esta prueba se basa en el __Teorema de Glivenko-Cantelli__, que establece que 
$$\mathbb{P}\left[ \lim_{n \rightarrow \infty} \sup_{x} | S_n(x) - F(x) | \right] = 0 $$
donde $F(x)$ es la función de distribución hipotética y $S_n(x)$ es la función de distribución empírica, la cual se define de la siguiente manera
$$ S_n(x) = \left \{ \begin{array}{ll} 0 & X < X_{(0)} \\ \frac{i}{n} & X_{(i)} \leq x < X_{(i+1)} \\ 1 & x \geq X_{(n)} \end{array} \right.$$

__Estadístico de prueba:__ $D = \sup_{x} | S_n(x) - F^*(x)|$.

__Regla de decisión:__ Rechazar $H_0$ al nivel de significancia $\alpha$ si $D > q_{1-\alpha}$.

__Nota 1:__ Dada la distribución hipotética para el contraste de hipótesis, si no se conocen los parámetros, entonces se deben utilizar los mejores estimadores insesgados para los parámetros de dicha distribución. Por ejemplo, si utilizamos que $F(x) \sim N(\mu, \sigma^2)$ pero no se nos especifican los valores de la media y la varianza, debemos utilizar $\mu = \overline{X}$ y $\sigma^2 = S^2$.

__Nota 2:__ Esta prueba también puede utilizarse para probar si dos muestras provienen de una misma distribución, es decir, 
$$H_0: F_X(x) = F_Y(y) \text{ vs } H_1: F_X(x) \not = F_Y(y)$$

__Nota 3:__ La prueba de KS solo sirve para distribuciones continuas.

## Pruebas de Normalidad

La distribución normal es una de las distribuciones más empleadas en probabilidad y estadística debido a sus propiedades y la importancia que tiene en dichos campos. Por lo tanto, no resulta sorprendente que esta distribución siempre sea candidata para ajustar los datos de interés. 

Existen múltiples pruebas para demostrar si una muestra aleatoria se distribuye normal. 
Algunas pruebas frecuentemente utilizadas para probar la normalidad son: __Lilliefors__ (caso particular de la prueba Kolmogorov-Smirnov), __Anderson-Darling__, __Cramer-von Mises__, __Shapiro-Wilk__, __Jarque-Bera__.

Para todas las pruebas mencionadas el contraste de hipótesis es 
$$H_0: \text{Los datos se distribuyen normal} \text{ vs } H_1:\text{Los datos no se distribuyen normal}$$

```r
miau <- rnorm(100, 34, 18)
hist(miau, probability = TRUE, col = "light blue")
lines(density(miau), col = "dark blue", lwd = 3)
```

## Prueba t

La prueba t se utiliza para probar si la media de dos muestras es la misma; es decir, si tenemos $X_1,...,X_n$ y $Y_1,...,Y_m$ muestras aleatorias con medias $\mu_X$ y $\mu_y$ respectivamente, 
$$H_0: \mu_X = \mu_Y \text{ vs } H_1: \mu_X \not = \mu_y $$

Para aplicar la prueba t se deben cumplir tres supuestos:
1. Ambas muestras deben distribuirse normal. 
2. Las muestras deben ser independientes entre sí. En caso de no ser independientes se utilizará una prueba t para muestras pareadas. 
3. La varianza de X debe ser igual a la varianza de Y.

Para validar los supuestos es necesario aplicar pruebas de normalidad a ambas muestras y una prueba de homocedasticidad o varianza constante; algunas pruebas para homocedasticidad son __Breusch-Pagan, Levene, fligner__, entre otras. 

__Nota:__ En caso de no cumplir alguno de los supuestos anteriores se recomienda utilizar la __prueba de Wilcoxon-Mann-Whitney__.

```r
x <- rnorm(50, 14.2, 15)
y <- rnorm(60, 11.4, 11.6)

var.test(x,y)  # No rechazamos la varianza constante

t.test(x,y, mu = 0, paired = FALSE, var.equal = TRUE, alternative = c("two.sided"))# mu = 0 medias iguales. 
                                                        # paired = FALSE las muestras no estan pareadas
                                                        #var.equal = TRUE la varianza es igual en ambas muestras
                                                    #alternative two.sided Medias distintas
```

## Prueba de Wilcoxon-Mann-Whitney

Prueba no paramétrica que sirve para comparar las medias de dos muestras aleatorias independientes. A diferencia de la prueba t, la prueba de Wilcoxon solo necesita muestras independientes. El contraste de hipótesis es igual al contraste de la prueba t. Esta prueba es menos potente que la prueba t, por lo que debe se sugiere utilizar la prueba t en caso de que los supuestos se cumplan.

```r
x <- rnorm(50, 14.2, 15)
y <- rnorm(60, 11.4, 11.6)
wilcox.test(x,y, mu = 0, paired = FALSE, alternative = c("two.sided"))
```

## Prueba t por pares

En algunas ocasiones tenemos más de dos muestras aleatorias y nos interesa saber si existen diferencias entre al menos un par. La __prueba t por pares__ es la generalización de la prueba t para más de dos muestras. Los supuestos son similares a la prueba t; sin embargo, para verificar la varianza constante entre todas las muestras, esto no debe hacerse dos a dos, sino basta con comprobar la varianza agrupada (se sugiere utilizar el análisis de varianza ANOVA).

__Ejemplo 1:__ Un biólogo revisa los datos "iris" y conjetura que el ancho del sépalo no es igual entre las tres especies registradas. Prueba su hipótesis bajo un nivel de significancia $\alpha = 0.05$.

```r
data("iris")
summary(iris)

aggregate(Sepal.Width ~ Species, data = iris, mean)  #Vemos la media por especie
aggregate(Sepal.Width ~ Species, data = iris, sd)   #Vemos la desv est por especie

ana <- aov(Sepal.Width ~ Species, data = iris)  #Tabla anova
summary(ana) #Existe varianza constante

pairwise.t.test(x = iris$Sepal.Width, g = iris$Species, 
                pool.sd = TRUE, paired = FALSE)    #Explicamos el ancho del sepalo por las especies
                                                    #pool.sd = TRUE ya que hay varianza constante

plot(TukeyHSD(ana))  #Graficamos la diferencia 2 a 2 de las medias
```
